```python
import pandas as pd

medicine = {"Name": ['Pantop', 'Asprin', 'Zoloft', 'Ultraset'], "Price":[100, 200, 400, 700], "Prescription_Required":[False, False, True, True]}

df = pd.DataFrame(medicine)

print(df)
```

```python
## Slicing
# Print the first two rows
print(df.head(2))
print('\n')
# Print the last two rows
print(df.tail(2))
```

```python
## Merge two tables

import pandas as pd

medicine_a = {"Name": ['Pantop', 'Asprin', 'Zoloft', 'Ultraset'], "Price":[100, 200, 400, 700], "Prescription_Required":[False, False, True, True], "index" : [100, 200, 300, 400]}
medicine_b = {"Name": ['Rivotril', 'Crocin', 'Limcee', 'Superspas'], "Price":[400, 500, 150, 100], "Prescription_Required":[True, False, False, False], "index" : [100, 200, 300, 400]}

dfa = pd.DataFrame(medicine_a)
dfb = pd.DataFrame(medicine_b)

print(dfa)
print('\n')
print(dfb)
print('\n')
mergex = pd.merge(dfa, dfb)
print(mergex)
print('\n')
```

```python
#Changing the index value

df = pd.DataFrame({"Day" :[1,2,3,4], "Visitors":[200,200,230,300], "Bounce_Rate": [20,45,60,10]})

df.set_index("Day", inplace = True)

print(df)
```

```python
# Plotting

import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")

df = pd.DataFrame({"Day" :[1,2,3,4], "Visitors":[200,200,230,300], "Bounce_Rate": [20,45,60,10]})

df.set_index("Day", inplace = True)

df.plot()
plt.show()

```

```python
## Concatenation

import pandas as pd

medicine_a = {"Name": ['Pantop', 'Asprin', 'Zoloft', 'Ultraset'], "Price":[100, 200, 400, 700], "Prescription_Required":[False, False, True, True], "index" : [100, 200, 300, 400]}
medicine_b = {"Name": ['Rivotril', 'Crocin', 'Limcee', 'Superspas'], "Price":[400, 500, 150, 100], "Prescription_Required":[True, False, False, False], "index" : [500, 600, 700, 800]}

dfa = pd.DataFrame(medicine_a)
dfb = pd.DataFrame(medicine_b)

concat = pd.concat([dfa, dfb])
print(concat)
```
